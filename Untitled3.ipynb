```python
import numpy as np
```

```python
x = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])
```

```python
def sigmoid(x, d = False): 
    if d: 
        return x * (1-x)
    return 1/(1+np.exp(-x))


def init_p(): 
    np.random.seed(42)
    
    w1 = np.random.randn(2, 4)*0.5
    b1 = np.zeros((1,4))

    w2 = np.random.randn(4,1) * 0.5 
    b2 = np.zeros((1,1))

    return w1, b1, w2, b2   

def get_loss(op, y, l_t): 
    e = 1e-8 
    if l_t == 'mse':
        return op - y
    elif l_t == 'bce': 
        return (op - y)/(op+e)*(1-op+e)

def train(x, y, lr =0.1, l_t='mse', epochs = 500): 
    w1, b1, w2, b2 = init_p()
    m = x.shape[0] 

    for _ in range(epochs): 
        z1 = np.dot(x,w1) + b1 
        a1 = sigmoid(z1) 
        z2 = np.dot(a1,w2) + b2
        a2 = sigmoid(z2)

        #backward pass 
        dz2 = get_loss(a2, y, l_t) * sigmoid(a2, True)
        dw2 = (1/m) * np.dot(a1.T, dz2) 
        db2 = (1/m) * np.sum(dz2, axis=0, keepdims=True)

        da1 = np.dot(dz2, w2.T) 
        dz1 = da1 * sigmoid(a1, True)
        dw1 = (1/m) * np.dot(x.T, dz1) 
        db1 = (1/m) * np.sum(dz1, axis=0, keepdims=True)

        w2 -= lr*dw2
        b2 -= lr*db2 
        b1 -= lr*db1 
        w1 -= lr*dw1

    return w1, b1, w2, b2 
         

```

```python
def get_acc(x,y,w1,b1,w2,b2):
    z1 = np.dot(x,w1) + b1 
    a1 = sigmoid(z1) 
    z2 = np.dot(a1, w2) + b2 
    a2 = sigmoid(z2) 
    return np.mean((a2 > 0.5) == y)*100 
```

```python
for l in [0.01, 0.1, 1.0]: 
    w1, b1, w2, b2 = train(x,y, lr=l, l_t='mse')
    acc = get_acc(x,y,w1,b1,w2,b2)
    print(f"LR= {l}, Acc: {acc:.1f}%")
```

```python
for l in ['mse', 'bce']: 
    w1, b1, w2, b2 = train(x,y, lr=0.1, l_t=l) 
    acc = get_acc(x,y,w1,b1,w2,b2) 
    print(f"Loss: {l}, Acc: {acc:.1f}%")
```
