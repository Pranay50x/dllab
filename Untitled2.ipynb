```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = np.zeros((300,300), dtype=np.uint8)
cv2.circle(img, (150,150), 75, 255, -1)
plt.imshow(img, cmap='gray')

```

```python
noisy = img.copy() 
noise = np.random.random(img.shape)
noisy[noise < 0.2] = 0 
noisy[noise > 0.98] = 255
plt.imshow(noisy, cmap='gray')
```
