```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = np.ones((300,300,3), np.uint8)*255 
cv2.rectangle(img, (100,100), (200, 200), (255,0,0), 2)
plt.imshow(img)
```

```python
c, angle, tx, ty = (150,150), 45, 50 , 30 
m_rot = cv2.getRotationMatrix2D(c, angle, 1.0) 
rot = cv2.warpAffine(img, m_rot, (300,300))
plt.imshow(rot)
```

```python
m_tran = np.float32([[1,0,tx], [0,1,ty]])
trans = cv2.warpAffine(img, m_tran, (300,300))
plt.imshow(trans)
```

```python
m_comb = m_rot.copy()
m_comb[:,2] += [tx, ty]
comb = cv2.warpAffine(img, m_comb, (300,300))
plt.imshow(comb)
```
