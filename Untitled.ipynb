```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = plt.imread('cliff-edge.jpg')
plt.imshow(img)
```

```python
r = img[:, : , 0]
g = img[:, : , 1]
b = img[:, : , 2]

gray = (.299*r + .587*g + .114*b).astype(np.uint8)
```

```python
plt.imshow(gray, cmap='gray')
```

```python
binary = np.where(gray > 127, 255, 0)
plt.imshow(binary, cmap='gray')
```
